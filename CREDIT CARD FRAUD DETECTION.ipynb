## TRAINING,TESTING,MODEL BUILDING ALONG WITH PREDICTION

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(x_train,y_train)
```

```python
y_predict=model1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,precision_score,recall_score,f1_score
```

```python
print("Classification Report:",classification_report(y_test,y_predict))
print("Confusion Matrix:",confusion_matrix(y_test,y_predict))
print("accuracy score:",accuracy_score(y_test,y_predict))
print("Precision score:",precision_score(y_test,y_predict))
print("recall score:",recall_score(y_test,y_predict))
print("f1_score:",f1_score(y_test,y_predict))
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
model2=DecisionTreeClassifier()
model2.fit(x_train,y_train)
```

```python
plot_tree(model2,fontsize=1,filled=True)
plt.show()
```

```python
y_predict=model2.predict(x_test)
```

```python
print("Classification Report:",classification_report(y_test,y_predict))
print("Confusion Matrix:",confusion_matrix(y_test,y_predict))
print("accuracy score:",accuracy_score(y_test,y_predict))
print("Precision score:",precision_score(y_test,y_predict))
print("recall score:",recall_score(y_test,y_predict))
print("f1_score:",f1_score(y_test,y_predict))
```

```python
from sklearn.ensemble import RandomForestClassifier
model3=RandomForestClassifier()
model3.fit(x_train,y_train)
```

```python
y_predict=model3.predict(x_test)
```

```python
print("Classification Report:",classification_report(y_test,y_predict))
print("Confusion Matrix:",confusion_matrix(y_test,y_predict))
print("accuracy score:",accuracy_score(y_test,y_predict))
print("Precision score:",precision_score(y_test,y_predict))
print("recall score:",recall_score(y_test,y_predict))
print("f1_score:",f1_score(y_test,y_predict))
```

```python
final_data=pd.DataFrame({'models':['model1','model2','model3'],
             "AC":[accuracy_score(y_test,y_predict)*100,
                   accuracy_score(y_test,y_predict)*100,
                   accuracy_score(y_test,y_predict)*100],
                "PC":[precision_score(y_test,y_predict)*100,
                      precision_score(y_test,y_predict)*100,
                      precision_score(y_test,y_predict)*100],
              "RS":[recall_score(y_test,y_predict)*100,
                    recall_score(y_test,y_predict)*100,
                    recall_score(y_test,y_predict)*100],
                 "F1":[f1_score(y_test,y_predict)*100,
                    f1_score(y_test,y_predict)*100,
                    f1_score(y_test,y_predict)*100],
             })
```

```python
final_data
```

```python
plt.subplot(2,2,1)
sns.barplot(x=final_data["models"],y=final_data["AC"])
plt.subplot(2,2,2)
sns.barplot(x=final_data["models"],y=final_data["PC"])
plt.subplot(2,2,3)
sns.barplot(x=final_data["models"],y=final_data["RS"])
plt.subplot(2,2,4)
sns.barplot(x=final_data["models"],y=final_data["F1"])
plt.show()
```
